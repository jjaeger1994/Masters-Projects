## Step 5: Filter Results and Build and Validate Models


```r
diabetes$Outcome <- ifelse(diabetes$Outcome == 1, "Diabetic", "NonDiabetic")
diabetes$Outcome <- factor(diabetes$Outcome)
```

```r
##Remove non-sensical values
valid_diabetes <- diabetes %>%
    filter(Glucose != 0 & BloodPressure != 0 & BMI != 0)
```

```r
head(valid_diabetes)
dim(valid_diabetes)
```

```r
diabetes_model <- valid_diabetes %>%
    select(-Glucose,-Insulin) %>%
    rename(DPF = DiabetesPedigreeFunction)

head(diabetes_model)
```

```r
set.seed(1)
size = round(dim(diabetes_model)[1]*0.8)

train_ind <- sample(seq_len(nrow(diabetes_model)), size=size)

train <- diabetes_model[train_ind,]
test <- diabetes_model[-train_ind, ]

head(test)
```

```r
##predict diabetes with all easily attainable clinical info
formula <- Outcome ~ .
```

<h3> Method </h3>
We will choose a tree because it will provide an easy to use model for clinicians that don't have access to computers able to run the algorithm.

```r
tree_model <- tree(formula, data = train)

```

```r
summary(tree_model)
```

```r
outcome_pred <- predict(tree_model, test[,-7], type="class")

print("Misclassified: ")
sum(outcome_pred != test$Outcome)
print("Out of: ")
dim(test)[1]
```

## Step 6: Visualize Results
