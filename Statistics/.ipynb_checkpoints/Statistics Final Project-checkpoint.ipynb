From my exploratory data analysis, I have been able to remove some insignificant variables and create new ones, modelVars will hold the variables we want.

```r
modelVars <- carPricing %>%
    select(fueltype,aspiration,carbody,drivewheel,
           enginelocation,enginetype,cylindernumber,fuelsystem,wheelbase,
           carlength,carwidth,curbweight,enginesize,boreratio,horsepower,citympg,highwaympg,citympg.2,highwaympg.2,
           curbweight.2,carlength.2,LogPrice)
```

When performing the linear model, there were a few problem variables causing issues with perfect multicollinearity

```r
modelVars %>%
    group_by(cylindernumber) %>%
    count()
```

```r
modelVars %>%
    group_by(fuelsystem) %>%
    count()
```

```r
table(modelVars$cylindernumber,modelVars$fuelsystem)
```

```r
table(modelVars$cylindernumber,modelVars$enginetype)
```

fuel system and cylinder number are problematic because they have a few categories with only one instance, 
enginetype can explain some of the variance explained by fuelsystem and cylinernumber, so we will keep it
and drop the others

```r
modelVars <- select(modelVars, -fuelsystem, -cylindernumber)
```

I will be performing a linear regression to determine how well I can predict the price of a car from the given variables.

Test/Train split. 

```r
set.seed(100)

split = sample.split(modelVars$LogPrice, SplitRatio = 0.7)

train <- subset(modelVars, split == TRUE)
test <- subset(modelVars, split == FALSE)
```

Forward selection of the data

```r
regfit.best <- regsubsets(LogPrice~.,data= modelVars, method="forward",nvmax=20)
reg.summary <- summary(regfit.best)
```

```r
reg.summary
```

I will be using BIC to determine the best model

```r
minbic = which.min(reg.summary$bic)
```

```r
plot(reg.summary$bic, xlab="Number of Variables", ylab="BIC", type='l')
points(minbic, reg.summary$bic[minbic], col="red", cex=2, pch=20)
```

```r
png(filename = "BICBest.png")
plot(reg.summary$bic, xlab="Number of Variables", ylab="BIC", type='l')
points(minbic, reg.summary$bic[minbic], col="red", cex=2, pch=20)
dev.off()
```

```r
plot(regfit.best, scale= "bic")
```

```r
png("bicvars.png")
plot(regfit.best, scale= "bic")
dev.off()
```

What features were used in the final model?

```r
coef(regfit.best, minbic)
```

Variance Explained by R squared:

```r
reg.summary$rsq[minbic]
```

get the features from the best model

```r
bestmodel <- names(coef(regfit.best,minbic))
```

Create a lm object for it's analytical features

```r
train_matrix <- select(as.data.frame(model.matrix(~.,data=train)),all_of(bestmodel),LogPrice)
```

```r
test_matrix <- select(as.data.frame(model.matrix(~.,data=test)),all_of(bestmodel),LogPrice)
```

```r
lm.model <- lm(LogPrice~.+0,data=train_matrix)
summary(lm.model)
```

The estimates provided by the lm model differ slightly than those generated by regsubset, but the regression still performs very well.

```r
options(repr.plot.width = 15, repr.plot.height = 15)
par(mfrow = c(2,2), oma = c(0, 0, 1.1, 0))
plot(lm.model)
```

```r
png(filename="lmresults.png")
options(repr.plot.width = 15, repr.plot.height = 15)
par(mfrow = c(2,2), oma = c(0, 0, 1.1, 0))
plot(lm.model)
dev.off()
```

I would like to test how well the assumptions of the linear model hold up.

```r
shapiro.test(residuals(lm.model))
```

Residual meet the normality assumption

```r
bptest(lm.model)
```

We must reject the null hypothesis that variance is the same amongst residuals

```r
dwtest(lm.model, alternative= "two.sided")
```

We can reject the null hypothesis that there is autocorrelation between the residuals.

Overall, the linear model works well, so I will test it against the test data to prove it's predictive capabilities

```r
train_residual <- lm.model$residuals
test_predict <- predict(lm.model, newdata = test_matrix)

test_residual <- test_predict - test_matrix$LogPrice

trainSSE <- sum(train_residual^2)
testSSE <- sum(test_residual^2)

trainRMSE <- sqrt(trainSSE/nrow(train_matrix))
testRMSE <- sqrt(testSSE/nrow(test_matrix))

trainRMSE
testRMSE
```

the two models have similar RMSE, so the model continues to have strong predictive ability on the test data.

```r
testSST <- sum((test_matrix$LogPrice - mean(test_matrix$LogPrice))^2)


TestR2 <- 1-(testSSE/testSST) 
paste("The test R-Squared Value is: ",TestR2)
```

The R-Squared value is also strong on the test data.
