```r

consumption <- read.csv("EnergyConsumption.csv")
electric <- read.csv("Electric Production.csv")
```

```r
library(dplyr)
library(stringr)
```

```r
head(consumption)
```

```r
consumption$year = as.integer(str_sub(consumption$YYYYMM,1,4))
consumption$month = as.integer(str_sub(consumption$YYYYMM,5,6))
consumption$Description = as.character(consumption$Description)

## A month of 13 represents the annual totals.
consumptionclean <- consumption %>%
    filter(month == 13) %>%
    filter(Column_Order %in% c(1,2,3,5,6,7,8,9,10)) %>%
    select(Value, Description, year) %>%
    mutate(Value = str_replace(.$Value,"Not Available","0")) %>%
    mutate(Description = str_extract(.$Description,"\\w+")) %>%
    mutate(Description = str_replace(.$Description,"Natural","Natural Gas")) 

consumptionclean$Value = as.numeric(consumptionclean$Value)

head(consumptionclean)
```

```r
ggplot() +
    geom_line(data= consumptionclean, aes(x=year, y=Value, color= Description)) +
    geom_vline(aes(xintercept= 2006), linetype= "dashed") + 
    geom_text(aes(2006, 41, label= "An Inconvenient Truth"), nudge_x = -13) +
    scale_color_brewer(type="qual", palette = "Paired") +
    labs(y= "Consumption Quadrillion BTU", x="", title= "Annual Energy Consumption by Type") +
    theme_minimal() +
    theme(panel.grid.major = element_blank(), panel.grid.minor = element_blank()) 
    
   
    
```

```r
png(filename="AnnualConsumption.png")

ggplot() +
    geom_line(data= consumptionclean, aes(x=year, y=Value, color= Description)) +
    geom_vline(aes(xintercept= 2006), linetype= "dashed") + 
    geom_text(aes(2006, 41, label= "An Inconvenient Truth"), nudge_x = -13) +
    scale_color_brewer(type="qual", palette = "Paired") +
    labs(y= "Consumption Quadrillion BTU", x="", title= "Annual Energy Consumption by Type") +
    theme_minimal() +
    theme(panel.grid.major = element_blank(), panel.grid.minor = element_blank()) 
    
dev.off()  
    
```

```r
DescGroup <- consumptionclean %>%
    group_by(Description) %>%
    mutate(percent_change = (Value - lag(Value))/Value *100)

head(DescGroup)
```

```r
ggplot(DescGroup) +
    geom_line(aes(x=year, y=percent_change, color= Description)) +
    geom_vline(aes(xintercept= 2006), linetype= "dashed") + 
    geom_text(aes(2006, 41, label= "An Inconvenient Truth"), nudge_x = -12) +
    scale_color_brewer(type="qual", palette = "Paired") +
    labs(y= "Yearly Percent Change", x="") +
    theme_minimal() +
    theme(panel.grid.major = element_blank(), panel.grid.minor = element_blank())
```

```r
consumptionTotals <- consumption %>%
    filter(month == 13) %>%
    filter(Column_Order %in% c(4,5,11)) %>%
    select(Value, Description, year) %>%
    mutate(Value = str_replace(.$Value,"Not Available","0")) %>%
    mutate(Description = str_replace(.$Description,"Total Fossil Fuels Consumption","Fossil Fuels")) %>%
    mutate(Description = str_replace(.$Description,"Total Renewable Energy Consumption","Renewables")) %>%
    mutate(Description = str_replace(.$Description,"Nuclear Electric Power Consumption","Nuclear")) %>%
    filter(year > 2006)

consumptionTotals$Value = as.numeric(consumptionTotals$Value)

head(consumptionTotals)
```

```r
ggplot(consumptionTotals) +
    geom_bar(aes(fill=Description, x=year, weight=Value), position="fill") +
    theme_minimal() +
    theme(panel.grid.major = element_blank(), panel.grid.minor = element_blank()) +
    labs(y="",x="", title="Renewable's and Fossil Fuel's Share of Consumed Energy") +
    scale_fill_brewer(type="qual") 
```

```r
png(filename = "AnnualShare.png")

ggplot(consumptionTotals) +
    geom_bar(aes(fill=Description, x=year, weight=Value), position="fill") +
    theme_minimal() +
    theme(panel.grid.major = element_blank(), panel.grid.minor = element_blank()) +
    labs(y="",x="", title="Renewable's and Fossil Fuel's Share of Consumed Energy") +
    scale_fill_brewer(type="qual")

dev.off()
```

```r
DescGroup <- consumptionTotals %>%
    group_by(Description) %>%
    mutate(BTUchange = Value - lag(Value))

head(DescGroup)
```

```r
ggplot(DescGroup) +
    geom_col(aes(x=year, y=BTUchange, fill= Description), position = "identity") + 
    scale_fill_brewer(type="qual") +
    labs(y= "Yearly Quadrillion BTU Change", x="", title= "Annual Change in Consumption") +
    geom_hline(aes(yintercept=0), color = "red") +
    theme_minimal() +
    theme(panel.grid.major = element_blank(), panel.grid.minor = element_blank(), legend.position = "none") +
    facet_wrap(~Description)
```

```r
png(filename = "DiffConsumption.png")

ggplot(DescGroup) +
    geom_col(aes(x=year, y=BTUchange, fill= Description), position = "identity") + 
    scale_fill_brewer(type="qual") +
    labs(y= "Yearly Quadrillion BTU Change", x="", title= "Annual Change in Consumption") +
    geom_hline(aes(yintercept=0), color = "red") +
    theme_minimal() +
    theme(panel.grid.major = element_blank(), panel.grid.minor = element_blank(), legend.position = "none") +
    facet_wrap(~Description)

dev.off()
```

```r
electric$year = as.integer(str_sub(electric$YYYYMM,1,4))
electric$month = as.integer(str_sub(electric$YYYYMM,5,6))

electricclean <- electric %>%
    filter(month == 13) %>%
    filter(Column_Order == 13) %>%
    mutate(Value = str_replace(.$Value,"Not Available","0")) %>%
    select(Value, year)

electricclean$Value = as.numeric(electricclean$Value)

head(electricclean)
```

```r
ggplot(electricclean) +
    geom_line(aes(x= year, y= Value), color="orange") +
    labs(x= "", y= "Electric Production Million Kilowatthours", title = "Annual Electric Production") +
    geom_vline(aes(xintercept= 2006), linetype= "dashed") + 
    geom_text(aes(2006, 1000000, label= "An Inconvenient Truth"), nudge_x = -12) +
    theme_minimal() +
    theme(panel.grid.major = element_blank(), panel.grid.minor = element_blank()) 
```
