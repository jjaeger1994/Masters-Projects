```python
#Import libraries and read in Salary data
import pandas as pd
from datetime import date
import matplotlib.pyplot as plt
import cpi
cpi.update()
dat = pd.read_csv("Salaries.csv")
dat
```

## Inflate salaries to 2016 value

```python
real_salary_list = []

for i in range(len(dat)):
    real_salary = cpi.inflate(dat.iloc[i].salary, dat.iloc[i].yearID, to=2016)
    real_salary = round(real_salary)
    real_salary_list.append(real_salary)
```

## Add Real Salary to dataset

```python
dat['Real Salary'] = real_salary_list
dat
```
