<h1 align="center"> Variable Removal From Pitching DataFrame</h1>

```python
pitchingdat = pitchingdat.drop(columns = ["IPouts", "BFP", "R", "ER", "IBB", "BK", "GF", "G"])
```

```python
pitchingdat.info()
```

<h1 align="center"> Combining Wins and Loses with a W/L ratio</h1>
<h3 align="center"> By Combining Winds and Loses we can protect and preserve data fidelity </h3>


```python
pitchingdat.sample(10)
```

<h3 align="center"> Getting a Ratio of Wins and Losses and then deleting both the W and L columns </h3>
<h3 align="center">Because there are some zero values that would result in inf values, I converted those numbers to 0, meaning they have no W/L ratio at this time </h3>

```python
pitchingdat["Win/Loss"] = pitchingdat.W.div(pitchingdat.L)

pitchingdat.loc[~np.isfinite(pitchingdat['Win/Loss']), 'Win/Loss'] = 0

pitchingdat.info()
```

```python
pitchingdat = pitchingdat.drop(columns = ["L", "W"])
```

```python
pitchingdat.info()
```
